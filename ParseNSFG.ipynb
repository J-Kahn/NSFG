# Format data in terms of years on birth control

Create panel data. There may be more efficient ways of doing this, but it seems to work fine.

```python
dates = pd.date_range('1/1/1999', periods=51, freq='M')
```

```python
def stack_nsfg(name, append = ""):
    
    stackd = dat[[name + i + append for i in names]].stack(dropna=False)
    
    stackd.index = pd.MultiIndex.from_product([dat['caseid'], dates], names=['case', 'month'])
    stackd.name  = name + append
    
    return stackd
```

```python
panelvars = [stack_nsfg('preg'), 
       stack_nsfg('want'),
       stack_nsfg('fail'),
       stack_nsfg('methhist','1'),
       stack_nsfg('methhist','2'),
       stack_nsfg('methhist','3'),
       stack_nsfg('methhist','4')]
```

```python
panl = pd.concat(panelvars, axis = 1)
```

Get dummies for method use

```python
panl["methhist1"] = panl["methhist1"].fillna("Not applicable")
panl["methhist2"] = panl["methhist2"].fillna("Not applicable")
panl["methhist3"] = panl["methhist3"].fillna("Not applicable")
panl["methhist4"] = panl["methhist4"].fillna("Not applicable")
```

```python
panl["methhist1"] = panl["methhist1"].map(str.strip)
panl["methhist2"] = panl["methhist2"].map(str.strip)
panl["methhist3"] = panl["methhist3"].map(str.strip)
panl["methhist4"] = panl["methhist4"].map(str.strip)
```

```python
panl['methhist'] = panl['methhist1'] + '|' + panl['methhist2'] + '|' + panl['methhist3'] + '|' + panl['methhist4']
```

```python
methdum = panl['methhist'].str.get_dummies()
```

## Use calendar years to do calculations.

Track period before pregnancy

```python
panly=panl['preg'].groupby(panl.major_axis.year)
```

```python
panl['b4preg'] = panl['preg']
    
```

```python
for lagm in range(1,12):
    panl['b4preg'] = panl['b4preg'] + panly.shift(lagm).fillna(0)
```

```python
panl.items
```

```python
panl
```

```python
ypreg = panl['preg'].resample('A', how=sum)
```

```python
yfail = (panl['fail']=='YES').resample('A', how=sum)
```

```python
def fail_percent(method):
    b4meth = panl[method]*(1-panl['b4preg']) + panl['b4preg']
    yb4meth = b4meth.resample('A', how=sum)
    ymeth = panl[method].resample('A', how=sum)
    yfb4meth = (yb4meth == 12) * (ymeth > 0)
    npreg = np.sum(np.sum((ypreg > 0) * yfb4meth))
    nmeth = np.sum(np.sum(yfb4meth))
    nfail = np.sum(np.sum((yfail > 0) * yfb4meth))
    return [nfail, 
            npreg, 
            nmeth, 
            float(nfail)/float(nmeth-npreg+nfail),
            float(npreg)/float(nmeth)]
```

```python
def constants(x):
    
    strx = x[0]
    strr = x[0]
    rowi = 0
    
    if type(x[0]) == float:
        if np.isnan(x[0]):
            strx = "NAN"
            strr = "NAN"
    
    rowi = 0
    
    for xi in x:
        
        if type(xi) == float:
            xi = "NAN"        
        
        if xi != strx:
            
            strx = xi
            
            strr = strr + ' -' + str(rowi) + '- ' strx
    
    return strr
```
